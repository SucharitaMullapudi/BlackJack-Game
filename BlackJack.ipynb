```python
# Global variables

import random

suits = ('Spades','Hearts','Diamond','Clubs')
numbers = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':1}
```

```python
class card_mix:
    def __init__(self,suits,numbers):
        self.suits = suits
        self.numbers = numbers
        self.values = values[numbers]
    def __str__(self):
        return self.numbers+' of '+ self.suits
```

```python
class card:
    def __init__(self):
        self.cards_list = []
        for suit in suits:
            for num in numbers:
                self.cards_list.append(card_mix(suit,num))
    def shuffle(self):
        random.shuffle(self.cards_list)
```

```python
class player_acct:
    
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
    def __str__(self):
        return f'The owner is {self.owner} with balance {self.balance}'
#     pass
    def deposit(self,amount_add):
        self.amount_add = amount_add
        self.balance = self.balance+amount_add
        print('Money accepted to player account,available bal : ',self.balance )
    def bet(self,amt_with):
        self.amt_with = amt_with
        if amt_with <= self.balance:
            print('withdraw accepted')
            self.balance = self.balance-amt_with
            print(self.balance)
        else:        
            print('funds unavailable')
```
